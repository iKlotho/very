```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blend_old = pd.read_excel("blend-eski.xlsx")
blend_new = pd.read_excel("blend-yeni.xlsx")
```

```python
blend_old = blend_old.dropna(how='all')
```

```python
def clean(df):
    df = df.dropna(how='all')
    df.loc[df.Destination=='Details','Destination'] = np.nan
    df["Vendor"] = df["Vendor"].fillna(1)
    df = df.fillna(method="ffill")
    df = df[df.Vendor!=1]
    return df
```
