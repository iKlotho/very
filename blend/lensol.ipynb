```python
def check_for_dests(df,check,vendor):
    df = df.dropna(how='all')
    df.loc[df.Destination=='Details','Destination'] = np.nan
    df = df.fillna(method="ffill")
    df_v = df[df.Vendor.isin(vendor)]
    df_v = df_v[df_v.Destination.isin(check)]
    return df_v
```
